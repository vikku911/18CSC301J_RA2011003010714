```python
import matplotlib.pyplot as plt

# data
players = ['V Kohli', 'S Dhawan', 'DA Warner', 'RG Sharma', 'SK Raina', 
           'AB de Villiers', 'MS Dhoni', 'CH Gayle', 'RV Uthappa', 'KD Karthik']
runs = [6624, 6244, 5881, 5879, 5528, 5162, 4978, 4965, 4952, 4376]

# create a bar chart
plt.bar(players, runs)

# set title and labels
plt.title('Top 10 players with most runs in IPL history')
plt.xlabel('Players')
plt.ylabel('Runs')

# rotate x-axis labels
plt.xticks(rotation=90)

# show the plot
plt.show()
```
